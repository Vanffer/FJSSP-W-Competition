```python
# 导入 CSV 与路径工具，用于读取实例属性表。
import csv
import os
```

```python
# 定义两类实例属性文件路径：FJSSP-W 与 FJSSP。
fjssp_w = r'/instances/InstanceData/FJSSP-W/data.csv'
fjssp = r'/instances/InstanceData/FJSSP/data.csv'
```

Select Problem

```python
# 选择当前要分析的属性文件（默认示例选择 FJSSP-W）。
path = os.getcwd() + fjssp_w
#path = fjssp
```

Read Data

```python
# 读取 CSV 所有行到内存，便于后续自定义筛选。
data = []
with open(path, 'r', newline='') as f:
    reader = csv.reader(f)
    for row in reader:
        data.append(row)
```

```python
# 分离表头和数据区，后续按表头字段名定位列索引。
header = data[0]
data = data[1:]
```

```python
# 定义筛选函数：每个字段设置上下界，返回满足全部条件的实例名。
def filter(bounds : dict):
    count = dict()
    for bound in bounds:
        index = header.index(bound)
        lb = bounds[bound][0]
        ub = bounds[bound][1]
        for instance in data:
            if float(instance[index]) >= lb and float(instance[index]) <= ub:
                if instance[0] not in count:
                    count[instance[0]] = 0
                count[instance[0]] += 1
    result = [instance for instance in count if count[instance] == len(bounds)] # check if the instance fits all set bounds
    return result
```

Available Attributes

```python
# 打印可筛选字段名，帮助构造过滤字典。
print(', '.join(header[1:]))
```

```python
# filter format: <attribute name>: (lower bound, upper bound)
example_filter = {
    'n_operations': (10, 50),
    'flexibility': (0.1, 0.2)
}
```

```python
# 执行筛选并返回实例列表。
filtered_instances = filter(example_filter)
```

```python
# 查看筛选结果数量。
len(filtered_instances)
```
